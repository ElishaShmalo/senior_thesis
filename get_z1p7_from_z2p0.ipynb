```python
import pandas as pd
import os
```

```python
def dec_to_str(num):
    return f"{num}".replace(".", "p")
```

```python
num_unit_cell_vals = [8, 16, 32, 64]
N_val = 4
a_vals = sorted(set([round(0.6 + 0.01 * i, 3) for i in range(0, 20)] 
          + [round(0.8 + 0.02 * i, 3) for i in range(0, 6)] +
          [round(0.7525 + 0.0025 * i, 4) for i in range(0, 8)] + [0.763]))

orig_z_val = 2.0
new_z_val = 1.7

orig_z_val_name = dec_to_str(orig_z_val)
new_z_val_name = dec_to_str(new_z_val)

num_samples = 1000


failed_a_vals = {}
for num_unit_cells in num_unit_cell_vals:
    failed_a_vals[num_unit_cells * N_val] = []

a_vals
```

```python


for a_val in a_vals:
    print(f"a_val: {a_val}")
    a_val_name = dec_to_str(a_val)
    for num_unit_cells in num_unit_cell_vals:
        L_val = num_unit_cells * N_val

        print(f"a_val: {a_val} | L_val {L_val}")

        stop_row = round(L_val ** new_z_val)
        sucsess = True

        path_name = f"data/spin_dists_per_time/N{N_val}/a{a_val_name}/IC1/L{L_val}/"
        for sample_num in range(1, num_samples+1):
            orig_file_name = f"N{N_val}_a{a_val_name}_IC1_L{L_val}_z{orig_z_val_name}_sample{sample_num}.csv"
            new_file_name = f"N{N_val}_a{a_val_name}_IC1_L{L_val}_z{new_z_val_name}_sample{sample_num}.csv"

            try:
                df = pd.read_csv(os.path.join(path_name, orig_file_name), nrows=stop_row)

                df.to_csv(os.path.join(path_name, new_file_name), index=False)
            except Exception as e:
                print(e)
                sucsess = False
            
            if sucsess == False:
                break
        
        if sucsess == False:
            print(f"Failed at: L = {L_val} | a = {a_val}")
            failed_a_vals[L_val].append(a_val)
            continue
```

```python
print(failed_a_vals)
```

```python
all_failed_a_vals = set(sum(list(failed_a_vals.values()), []))
print(all_failed_a_vals)

sucessful_a_vals = [a_val for a_val in a_vals if a_val not in all_failed_a_vals]

print(sucessful_a_vals)
```
