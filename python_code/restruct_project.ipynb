```python
# Make N folders
import os


Ns = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for data_kind in os.listdir(f"../{parent_folder}"):
    if data_kind == ".DS_Store":
        continue
    os.mkdir(f"../{parent_folder}/{data_kind}/SeveralNs")
    for N_val in Ns:
        os.mkdir(f"../{parent_folder}/{data_kind}/N{N_val}")
```

```python
# Make a folders
import os
a_vals = [i/100 for i in range(50, 101, 1)]

for data_kind in os.listdir(f"../{parent_folder}"):
    if data_kind == ".DS_Store":
        continue
    for N_val in os.listdir(f"../{parent_folder}/{data_kind}"):
        if N_val == ".DS_Store":
            continue
        os.mkdir(f"../{parent_folder}/{data_kind}/{N_val}/SeveralAs")
        for a_val in a_vals:
            os.mkdir(f"../{parent_folder}/{data_kind}/{N_val}/a_{str(a_val).replace(".", "p")}")
```

```python
# num_init_conds = [200]
num_init_conds = [1, 10, 50, 100, 500]
# Make init_cond folders

for data_kind in os.listdir(f"../{parent_folder}"):
    if data_kind == ".DS_Store":
        continue
    for N_val in os.listdir(f"../{parent_folder}/{data_kind}"):
        if N_val == ".DS_Store":
            continue
        for a_folder in os.listdir(f"../{parent_folder}/{data_kind}/{N_val}"):
            if a_folder == ".DS_Store":
                continue
            # os.mkdir(f"../{parent_folder}/{data_kind}/{N_val}/{a_folder}/SeveralIC")
            for num_init_cond in num_init_conds:
                os.mkdir(f"../{parent_folder}/{data_kind}/{N_val}/{a_folder}/IC{num_init_cond}")
```

```python
# Making folder for diffrent L_Vals
L_vals = [128, 256, 64]

def  get_nearest(N, L):
    """Returns the nearest integer to L that is a multiple of N"""
    return round(L/N) * N

# Make init_cond folders
for data_kind in os.listdir(f"../{parent_folder}"):
    if data_kind == ".DS_Store":
        continue
    for N_val in os.listdir(f"../{parent_folder}/{data_kind}"):
        if N_val == ".DS_Store":
            continue
        for a_folder in os.listdir(f"../{parent_folder}/{data_kind}/{N_val}"):
            if a_folder == ".DS_Store":
                continue
            for IC_folder in os.listdir(f"../{parent_folder}/{data_kind}/{N_val}/{a_folder}"):
                if IC_folder == ".DS_Store":
                    continue
                os.mkdir(f"../{parent_folder}/{data_kind}/{N_val}/{a_folder}/{IC_folder}/SeveralLs")
                if not (N_val == "SeveralNs"):
                    os.mkdir(f"../{parent_folder}/{data_kind}/{N_val}/{a_folder}/{IC_folder}/L{get_nearest(int(N_val[1:]), L_val)}")
```

```python
import os
import shutil

parent_folder = "data"

data_type = "delta_evolved_spins"

for N_val in os.listdir(f"../{parent_folder}/{data_type}"):
    if N_val == ".DS_Store":
            continue
    for filename in os.listdir(f"../{parent_folder}/{data_type}/{N_val}"):
        if filename == ".DS_Store":
            continue
        first = filename.index("_")
        second = filename.index("_", first + 1)
        third = filename.find("_", second + 1)
        fourth = filename.find("_", third + 1)

        N_folder = filename[0: first]
        a_val_str = str(round(float(filename[first+1: second][1:].replace("p", ".")), 3)).replace(".", "p")
        a_folder = "a" + str(round(float(a_val_str[1:].replace("p", ".")), 2)).replace(".", "p")
        init_cond_folder = filename[second+1: third]
        L_folder = filename[third+1:fourth]
        suffix = filename[fourth:]
        try:
            shutil.copy(
                f"../{parent_folder}/{data_type}/{N_val}/{filename}",
                f"../{parent_folder}2/{data_type}/{N_folder}/{a_folder}/{init_cond_folder}/{L_folder}/{filename}")
        except:
            print(filename)
```
